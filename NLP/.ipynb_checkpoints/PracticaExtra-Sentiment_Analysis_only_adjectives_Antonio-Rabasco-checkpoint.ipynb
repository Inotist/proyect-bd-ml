# Extra - Sentiment Analysis with POS tagging

Me quedé con el gusanillo de probar un modelo de análisis del sentimiento que únicamente tenga en cuenta los adjetivos que aparecen en el corpus y descarte el resto de palabras, y como he visto que voy bien de tiempo voy a hacer el ejercicio.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier

from keras.models import Sequential
from keras.layers import Dense, Dropout, LSTM

from collections import Counter

from gensim.models import Word2Vec

import spacy
spacy_nlp = spacy.load('en_core_web_sm')
```

```python
aiv5 = pd.read_csv("Amazon_Instant_Video_5.csv")
a5 = pd.read_csv("Automotive_5.csv")
dm5 = pd.read_csv("Digital_Music_5.csv")
mi5 = pd.read_csv("Musical_Instruments_5.csv")
plg5 = pd.read_csv("Patio_Lawn_and_Garden_5.csv")

reviews = pd.concat([aiv5,a5,dm5,mi5,plg5], ignore_index=True).dropna(subset=['reviewText','overall']).sample(frac=1)
reviews.head()
```

Esta vez solo quiero quedarme con los adjetivos, no necesito ningún preprocesado adicional.

```python
def preprocess_texts(texts, **kwargs):
    
    processed_texts = [] # Aquí me guardaré los textos procesados en forma de lista
    
    for text in texts:
        
        # Si decido quedarme solo con adjetivos
        if 'only_adjectives' in kwargs:
            adjectives = kwargs['only_adjectives'](text)
            tokens = [token.text for token in adjectives if token.pos_ == 'ADJ']
        
        processed_texts.append(tokens)
        
    return processed_texts
```

Categorizo igual que en el ejercicio original

```python
reviews = reviews[['reviewText','overall']]
    
reviews.overall = reviews.overall.apply(lambda x: 'very_good' if x == 5 else 'good' if x > 3 else 'not_so_good')
```

Primero sacaré los resultados usando el tf-idf y un random forest y después con word2vec haciendo uso de deep learning.

```python
X_train, X_test, y_train, y_test = train_test_split(
    reviews.reviewText,
    reviews.overall,
    train_size=0.80,
    test_size=0.20,
)

train_cv = X_train.copy()
for ix, processed_text in zip(X_train.index,
                              preprocess_texts(X_train, only_adjectives=spacy_nlp)):
    train_cv[ix] = " ".join(processed_text)

test_cv = X_test.copy()
for ix, processed_text in zip(X_test.index,
                              preprocess_texts(X_test, only_adjectives=spacy_nlp)):
    test_cv[ix] = " ".join(processed_text)
    
train_w2v = preprocess_texts(X_train, only_adjectives=spacy_nlp)
test_w2v = preprocess_texts(X_test, only_adjectives=spacy_nlp)
```

```python
cv = TfidfVectorizer(
    max_df=1.0,
    min_df=0.01,
    max_features=2500,
    ngram_range=(2,3)
)

train_cv = cv.fit_transform(train_cv)
test_cv = cv.transform(test_cv)
```

```python
maxDepth = range(5,15)

gridRFC = GridSearchCV(RandomForestClassifier(n_estimators=200), param_grid={'max_depth': maxDepth}, cv=30)
gridRFC.fit(train_cv, y_train)

print("best mean cross-validation score: {:.3f}".format(gridRFC.best_score_))
print("best parameters: {}".format(gridRFC.best_params_))

scores = np.array(gridRFC.cv_results_['mean_test_score'])
plt.plot(maxDepth,scores,'-o')
plt.xlabel('max_depth')
plt.ylabel('30-fold ACC')

plt.show()
```

```python
maxDepthOptimo = gridRFC.best_params_['max_depth']

RFC = RandomForestClassifier(max_depth=maxDepthOptimo,n_estimators=200).fit(train_cv, y_train)

test_predict = RFC.predict(test_cv)

print('Confussion matrix:\n{}'.format(confusion_matrix(y_test, test_predict)))
print('\nClassification report:\n{}'.format(classification_report(y_test, test_predict)))
print('Accuracy score:{}'.format(accuracy_score(y_test, test_predict)))
```

Veamos si con word2vec mejoran los resultados

```python
w2v_sg = Word2Vec(sg=1, size=300, min_count=3, window=3, hs=0, negative=20, workers=4)

w2v_sg.build_vocab(train_w2v)

w2v_sg.train(sentences=train_w2v, total_examples=w2v_sg.corpus_count, epochs=50)
```

```python
train_vector = np.zeros((len(train_w2v), 40, 300))

for i1 in range(len(train_w2v)):
    for i2 in range(200):
        try: train_vector[i1][i2] = w2v_sg.wv[train_w2v[i1][i2]]
        except: pass
        
test_vector = np.zeros((len(test_w2v), 40, 300))

for i1 in range(len(test_w2v)):
    for i2 in range(200):
        try: test_vector[i1][i2] = w2v_sg.wv[test_w2v[i1][i2]]
        except: pass
```

```python
y_train_int = y_train.apply(lambda x: 2 if x == 'very_good' else 1 if x == 'good' else 0)
y_test_int = y_test.apply(lambda x: 2 if x == 'very_good' else 1 if x == 'good' else 0)

model = Sequential()
model.add(LSTM(512, input_shape=(train_vector.shape[1:]), return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(256, return_sequences=True))
model.add(Dropout(0.1))
model.add(LSTM(256))
model.add(Dense(256, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(3, activation='softmax'))

model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(train_vector, y_train_int, validation_split=0.1, epochs=20, batch_size=1000)
```

```python
predict = model.predict(test_vector)
test_predict = [max(range(len(probs)), key=probs.__getitem__) for probs in predict]

print('Confussion matrix:\n{}'.format(confusion_matrix(y_test_int, test_predict)))
print('\nClassification report:\n{}'.format(classification_report(y_test_int, test_predict)))
print('Accuracy score:{}'.format(accuracy_score(y_test_int, test_predict)))
```

Usando word2vec el resultado no es muy distinto a lo que tenía en el ejercicio original, aunque sí que parece empeorar un poco.

```python
predictions = []
for pred in test_predict:
    if pred == 2: predictions.append('very_good')
    if pred == 1: predictions.append('good')
    if pred == 0: predictions.append('not_so_good')

print('Predicciones              Real\n-------------------------------------')
for cat in ('very_good','good','not_so_good'):
    print(f'{cat:>11}: {Counter(predictions)[cat]:<10} {cat:>11}: {Counter(y_test)[cat]}')
print('_____________________________________')
for pred,real in zip(predictions,y_test):
    print(f'{pred:>12}        {real:>15}')
```
